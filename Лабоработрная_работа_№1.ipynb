```python
#16
path ='Product Gallery.xlsx'
xls = pd.ExcelFile(path)
xls.sheet_names
df = pd.read_excel(xls, sheet_name='Лист1')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

categories = df['Category'].unique()

num_categories = len(categories)

fig, axes = plt.subplots(num_categories, 1, figsize=(10, 5 * num_categories))

for i, category in enumerate(categories):
    category_data = df[df['Category'] == category]

    category_data = category_data.sort_values(by='Quantity')

    axes[i].plot(category_data['Quantity'], category_data['Profit'], marker='o', label=category, color='orange')

    axes[i].set_title(f'Зависимость прибыли от количества товаров для категории: {category}')
    axes[i].set_xlabel('Количество проданных товаров')
    axes[i].set_ylabel('Прибыль')
    axes[i].grid(True)
    axes[i].legend()

plt.tight_layout()

plt.show()
```

```python
#17
customer_price_stats = df.groupby('Customer')['UnitPrice (Products)'].agg(['mean', 'median']).reset_index()


customer_with_max_mean_price = customer_price_stats.loc[customer_price_stats['mean'].idxmax()]


print("Средняя и медианная стоимость товаров для каждого клиента:")
print(customer_price_stats)

print("\nКлиент с наибольшей средней стоимостью:")
print(customer_with_max_mean_price)
```

```python
#18

df['OrderDate'] = pd.to_datetime(df['OrderDate'])

df['YearMonth'] = df['OrderDate'].dt.to_period('M')

monthly_sales = df.groupby('YearMonth')['Sales'].sum().reset_index()

monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', linestyle='-', color='black')

plt.title('Продажи по месяцам')
plt.xlabel('Дата')
plt.ylabel('Продажи')
plt.grid(True)

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#19

company_customers = df.groupby('Customer Company')['Customer'].nunique().reset_index()

company_customers.columns = ['Company', 'Unique Customers']

company_with_max_customers = company_customers.loc[company_customers['Unique Customers'].idxmax()]

print("Количество уникальных клиентов для каждой компании:")
print(company_customers)

print("\nКомпания с наибольшим количеством уникальных клиентов:")
print(company_with_max_customers)
```

```python
#20
category_product_sales = df.groupby(['Category', 'Product'])['Quantity'].sum().reset_index()

top_products_by_category = category_product_sales.loc[category_product_sales.groupby('Category')['Quantity'].idxmax()]

print("Товары с наибольшим количеством продаж в каждой категории:")
print(top_products_by_category)
```

```python
# 22
df[10] = pd.to_datetime(df[10], errors='coerce')

dater = df[10].max()

bulletproof = dater - pd.DateOffset(months=6)
filtered_data = df[(df[10] >= bulletproof) & (df[10] <= dater)]
top = filtered_data.groupby(2)[28].sum().reset_index()

top.columns = ['City', 'TotalSales']

top.sort_values(by='TotalSales', ascending=False).head(3)
```

```python
#23

average_profit_by_category = df.groupby('Category')['Profit'].transform('mean')

df['Profit Exceeding Average'] = df['Profit'] - average_profit_by_category

top_5_exceeding_profit = df.nlargest(5, 'Profit Exceeding Average')

print("Топ-5 записей с наибольшим превышением прибыли над средней для категории:")
print(top_5_exceeding_profit[['Category', 'Product', 'Profit', 'Profit Exceeding Average']])
```
